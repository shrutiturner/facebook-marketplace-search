Not sure why trying to import the data from the csv file straight into a dataframe is causing errors, but a workaround is required.

```python
with open('Products.csv') as file:
    csv_reader = csv.reader(file)
    header = []
    header = next(csv_reader)
    
    rows = []
    for row in csv_reader:
        rows.append(row)
```

```python
header
```

```python
rows[0:2]
```

```python
df = pd.DataFrame(rows, columns=header)
df.head()
```

Now we know a dataframe can be created, it's worth checking that all the rows of data have been imported.

```python
print(len(rows), len(df))
```
